```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set(style='ticks')
```

```python
np.random.seed(1234)
x = np.random.normal(loc=0.0, scale=1.0, size=1000)
y = np.random.normal(loc=0.0, scale=1.0, size=1000)
```

```python
mask = (x < 3) & (x > -3) & (y < 3) & (y > -3)
x = x[mask]
y = y[mask]
```

```python
mask = y < x

x_1 = x[mask]
y_1 = y[mask]

x_2 = x[~mask]
y_2 = y[~mask]
```

```python
sns.distplot(x_1, label='set 1')
sns.distplot(x_2, label='set 2')
plt.xlabel('x')
plt.legend()
```

```python
sns.distplot(y_1, label='set 1')
sns.distplot(y_2, label='set 2')
plt.xlabel('y')
plt.legend()
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x_1, y_1, label='set 1')
sns.scatterplot(x_2, y_2, label='set 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
