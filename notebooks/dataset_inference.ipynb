## Flag changes with magnitude

```python
def changes_with_mag(data, columns):
    step = 0.5
    df = pd.DataFrame(columns=columns)
    for flag in df.columns:
        for max_mag in np.arange(20, 27, step):
            mean_val = data.loc[(data['MAG_GAAP_r'] < max_mag) & (data['MAG_GAAP_r'] > max_mag - 1), flag].mean()
            df.loc['r {} - {}'.format(max_mag - step, max_mag), flag] = '{:.4f}'.format(mean_val)
    display(df)
```

```python
columns = ['Flag_1', 'Flag_2', 'IMAFLAGS_ISO_1', 'MASK_2', 'MASK_13']
changes_with_mag(kids, columns)
```

```python
changes_with_mag(kids_cs, columns)
```

```python
changes_with_mag(kids_x_sdss, columns)
```
